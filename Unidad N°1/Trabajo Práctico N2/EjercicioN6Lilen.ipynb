## Simulación de un Controlador Proporcional para Regular la Velocidad de un Motor en RPM

En este código, se implementa un controlador proporcional en Python para regular la velocidad de un motor simulado en revoluciones por minuto (RPM). El objetivo es ajustar la velocidad del motor para que se acerque lo más posible a una velocidad deseada predefinida en RPM.

El controlador proporcional calcula la diferencia entre la velocidad deseada y la velocidad actual del motor, y luego aplica una corrección proporcional a esta diferencia para ajustar la velocidad actual hacia la deseada.

En esta simulación, se establece una velocidad deseada de 30 RPM. La ganancia proporcional del controlador (kp) se ha fijado en 0.5. La simulación se realiza durante un período de 10 segundos con un intervalo de tiempo de 0.1 segundos.

Se grafica tanto la velocidad deseada como la velocidad actual del motor en función del tiempo para visualizar el comportamiento del controlador proporcional.






```python
import matplotlib.pyplot as plt  # Importamos la librería matplotlib para graficar
import numpy as np  # Importamos la librería numpy para trabajar con arreglos numéricos

class ControladorProporcional:
    def __init__(self, velocidad_deseada, kp):
        self.velocidad_deseada = velocidad_deseada  # Velocidad deseada que queremos alcanzar en RPM
        self.kp = kp  # Ganancia proporcional del controlador
        self.velocidad_actual = 0  # Velocidad actual del motor (inicializada en 0 RPM)

    def actualizar(self, dt):
        error = self.velocidad_deseada - self.velocidad_actual  # Calculamos el error entre la velocidad deseada y la actual
        correccion = self.kp * error * dt  # Calculamos la corrección proporcional a aplicar
        self.velocidad_actual += correccion  # Actualizamos la velocidad actual del motor con la corrección

    def obtener_velocidad_actual(self):
        return self.velocidad_actual  # Devuelve la velocidad actual del motor

# Parámetros
velocidad_deseada = 30  # Velocidad que queremos alcanzar en RPM
kp = 0.5  # Ganancia proporcional del controlador
tiempo_total = 10  # Duración total de la simulación en segundos
dt = 0.1  # Intervalo de tiempo en segundos para la simulación

# Crear el controlador proporcional
controlador = ControladorProporcional(velocidad_deseada, kp)

# Simulación
num_pasos = int(tiempo_total / dt)  # Calculamos el número de pasos de la simulación
tiempo = np.linspace(0, tiempo_total, num_pasos)  # Creamos un arreglo de tiempo de 0 a tiempo_total
velocidades_deseadas = np.full_like(tiempo, velocidad_deseada)  # Creamos un arreglo de la velocidad deseada en todo el tiempo
velocidades_actuales = []  # Lista para almacenar la velocidad actual en cada paso de la simulación

# Bucle de simulación
for t in tiempo:
    controlador.actualizar(dt)  # Actualizamos la velocidad actual del motor
    velocidad_actual = controlador.obtener_velocidad_actual()  # Obtenemos la velocidad actual del motor
    velocidades_actuales.append(velocidad_actual)  # Agregamos la velocidad actual a la lista

# Gráfica
plt.plot(tiempo, velocidades_deseadas, label='Velocidad Deseada (RPM)')  # Graficamos la velocidad deseada en función del tiempo
plt.plot(tiempo, velocidades_actuales, label='Velocidad Actual (RPM)')  # Graficamos la velocidad actual en función del tiempo
plt.xlabel('Tiempo (s)')  # Etiqueta del eje x
plt.ylabel('Velocidad (RPM)')  # Etiqueta del eje y
plt.title('Controlador Proporcional para Motor en RPM')  # Título del gráfico
plt.legend()  # Mostramos la leyenda
plt.grid(True)  # Mostramos la cuadrícula en el gráfico
plt.show()  # Mostramos el gráfico
```


**Analizar su comportamiento y respuesta a diferentes entradas.**

En un controlador proporcional, la corrección que se aplica a la salida del sistema (en este caso, la velocidad del motor) es proporcional al error entre la salida deseada y la salida actual. Esto permite ajustar dinámicamente la velocidad del motor para que se acerque a la velocidad deseada a lo largo del tiempo.
correccion = self.kp * error * dt  # en esta seccion de código Calculamos la corrección proporcional a aplicar.


*   error: es la diferencia entre la velocidad deseada y la velocidad actual del motor en un momento dado. Si error es positivo, significa que la velocidad actual es menor que la deseada, y si es negativo, la velocidad actual es mayor que la deseada.
*   kp: es la ganancia proporcional del controlador. Este valor determina la sensibilidad del controlador a la magnitud del error. Un valor más alto de kp hará que el controlador reaccione más fuertemente ante el error, mientras que un valor más bajo hará que la corrección sea más suave.
*   dt: es el intervalo de tiempo entre cada paso de la simulación. Multiplicamos el error por dt para ajustar la magnitud de la corrección en función del tiempo transcurrido desde el último cálculo. Esto asegura que la corrección sea proporcional tanto al error como al tiempo, lo que proporciona una respuesta dinámica adecuada del controlador a cambios en el sistema.
