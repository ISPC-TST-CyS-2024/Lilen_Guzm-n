**Crear un programa en Python utilizando la biblioteca Matplotlib para graficar
la respuesta al escalón de un sistema de control dinámico en lazo cerrado.
Analizar la respuesta en función de los parámetros del sistema.**

En un sistema de control dinámico en lazo cerrado, la respuesta al escalón es la salida del sistema cuando se le aplica una entrada escalón, es decir, una entrada que cambia abruptamente de un valor constante a otro. La respuesta al escalón es una herramienta importante para analizar el rendimiento del sistema de control, ya que proporciona información sobre la velocidad de respuesta del sistema, la estabilidad y el error en estado estacionario.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los parámetros
velocidad_deseada = 150  # Velocidad deseada del motor en RPM
velocidad_inicial = 0  # Velocidad inicial del motor en RPM
kp = 1.0  # Ganancia proporcional del controlador

# Definir los parámetros del escalón
escalón_inicio = 5  # Tiempo de inicio del escalón en segundos
escalón_duracion = 10  # Duración del escalón en segundos
velocidad_escalón = 50  # Magnitud del escalón en RPM  (define cuánto aumentará la velocidad del motor cuando se aplique el escalón).

# Definir los parámetros de simulación
tiempo_total = 20  # Duración total de la simulación en segundos
dt = 0.1  # Intervalo de tiempo en segundos para la simulación

# Crear vector de tiempo
tiempo = np.arange(0, tiempo_total, dt)

# Inicializar vector de velocidad del motor
velocidad_motor = np.zeros_like(tiempo)
velocidad_motor[0] = velocidad_inicial

# Simular sistema en lazo cerrado
for i in range(1, len(tiempo)):
    error = velocidad_deseada - velocidad_motor[i - 1]  # Calcular el error de velocidad
    correccion = kp * error  # Aplicar la corrección proporcional
    velocidad_motor[i] = velocidad_motor[i - 1] + correccion  # Actualizar la velocidad del motor

# Aplicar el escalón al sistema
escalón_puntos = int(escalón_duracion / dt)  # Calcular número de puntos afectados por el escalón
velocidad_motor[int(escalón_inicio / dt):int((escalón_inicio + escalón_duracion) / dt)] += velocidad_escalón# Aplicar el escalón.
                                                              #aplica un escalón en la velocidad del motor durante un intervalo de tiempo específico.
                                                              #Calcula los índices donde comienza y termina el escalón en el vector de tiempo,
                                                              #[int(escalón_inicio / dt):int((escalón_inicio + escalón_duracion) / dt)]
                                                              #luego actualiza el vector de velocidad del motor sumando la magnitud del escalón durante
                                                              #ese intervalo de tiempo.

# Graficar la velocidad del motor y el escalón
plt.plot(tiempo, velocidad_motor, label='Velocidad del Motor')
plt.plot([escalón_inicio, escalón_inicio + escalón_duracion], [velocidad_escalón, velocidad_escalón], '--', label='Escalón', color='red')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (RPM)')
plt.title('Respuesta del Motor al Escalón con Control de Lazo Cerrado')
plt.legend()
plt.grid(True)
plt.show()
```

**Analisis de la respuesta en función de los parámetros del sistema.**

*   Aumentar la ganancia proporcional (kp) hará que el sistema responda más rápido al error entre la velocidad deseada y la velocidad actual del motor. Sin embargo, un valor de kp demasiado alto puede provocar oscilaciones o inestabilidad en el sistema.
*   Una velocidad deseada más alta requerirá que el sistema alcance velocidades más altas y puede afectar el tiempo necesario para estabilizarse.
*   Una velocidad inicial más alta puede requerir menos tiempo para alcanzar la velocidad deseada, mientras que una velocidad inicial baja puede requerir más tiempo y energía para ajustarse al valor deseado.
*   La duración y la magnitud del escalón determinan cómo y cuánto cambiará la velocidad del motor cuando se aplique el escalón. Un escalón con una duración más larga o una magnitud más alta resultará en un cambio más significativo en la velocidad del motor y puede afectar la estabilidad del sistema.
