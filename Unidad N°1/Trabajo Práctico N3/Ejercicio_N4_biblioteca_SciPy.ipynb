**Crear un programa en Python utilizando la biblioteca SciPy para analizar la
respuesta en frecuencia de un sistema lineal invariante en el tiempo. Graficar
la respuesta en magnitud y fase.**



La respuesta en frecuencia de un sistema lineal invariante en el tiempo (LTI) es una medida de cómo el sistema responde a diferentes frecuencias de entrada. En términos simples, nos da una idea de cómo el sistema atenúa o amplifica diferentes frecuencias de una señal de entrada.

Un sistema lineal invariante en el tiempo puede representarse mediante su función de transferencia, que describe cómo la amplitud y la fase de las frecuencias de entrada se modifican al pasar a través del sistema.

La respuesta en frecuencia se puede caracterizar mediante dos componentes principales:

Magnitud de la respuesta en frecuencia: Indica cuánto amplifica o atenúa el sistema las diferentes frecuencias de la señal de entrada. Se suele expresar en decibelios (dB) o en escala lineal.
Fase de la respuesta en frecuencia: Indica cuánto se retrasa o adelanta la señal de salida en relación con la señal de entrada para diferentes frecuencias. Se mide en grados.
La respuesta en frecuencia es una herramienta fundamental en el análisis y diseño de sistemas de control, filtrado de señales, procesamiento de señales de audio y comunicaciones, entre otros campos. Permite entender cómo un sistema responde a diferentes frecuencias y ajustar su comportamiento según las necesidades específicas de una aplicación.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal #Proporciona herramientas para el análisis de señales y sistemas, incluyendo el cálculo de la respuesta en frecuencia.


# Definir la función de transferencia del sistema
#Aquí definimos la función de transferencia del sistema,
#especificando los coeficientes del numerador y del denominador del polinomio.

numerator = [1]  # Coeficientes del numerador del polinomio
denominator = [1, 2, 1]  # Coeficientes del denominador del polinomio
sys = signal.TransferFunction(numerator, denominator)

# Calcular la respuesta en frecuencia
#Utilizamos la función signal.bode para calcular la respuesta en frecuencia del sistema.
#Esta función calcula la frecuencia angular (w), la magnitud (mag),
#y la fase (phase) de la respuesta en frecuencia del sistema.

w, mag, phase = signal.bode(sys)

# Graficar la respuesta en magnitud
plt.figure()
plt.semilogx(w, mag)
plt.title('Respuesta en Magnitud')
plt.xlabel('Frecuencia [rad/s]')
plt.ylabel('Magnitud [dB]')
plt.grid(True)

# Graficar la respuesta en fase
plt.figure()
plt.semilogx(w, phase)
plt.title('Respuesta en Fase')
plt.xlabel('Frecuencia [rad/s]')
plt.ylabel('Fase [grados]')
plt.grid(True)

plt.show()

```
